```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate a dataset
np.random.seed(0) # set a random seed for reproducibility
months = pd.date_range(start ='2023-01-01', periods=12, freq='ME').strftime('%B') #create a list of month names
sales = np.random.randint(200,1000, size=12) # Genarate random sales data between 200 and 1000
profit = np.random.randint(20,100, size=12) #Generate random profit data between 20 and 100

# create a DataFrame
df = pd.DataFrame({
    'Month':months,
    'Sales':sales,
    'profit':profit,
})
df
```

```python
plt.figure(figsize=(14,7)) # set the figure size
bars=plt.bar(df['Month'], df['Sales'],color='teal',label='Sales') # create the bar chart with teal bars
plt.xlabel('Month') # set the x-axis label
plt.ylabel('Sales') # set the y-axis label
plt.title('Monthly Sales') # set the title of the chart
plt.xticks(rotation=45) # Rotate the x-axis labels for better readability
plt.grid(axis='y', linestyle='--',alpha=0.7) # Add horizontal gridlines with a dashed line style
plt.legend() # Add a legend

for bar in bars: # Add value labels on top of bars
    yval = bar.get_height() # get the height of each bar
    plt.text(bar.get_x() + bar.get_width()/2-0.1,yval+20,yval) # place the text annotation above each bar
    
plt.show() # Display the bar chart
```

```python
plt.figure(figsize=(14,7)) # set the figure size
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='teal', label='Sales') # create the bar chart with teal bars
plt.plot(df['Month'], df['profit'], marker='o',linestyle='-',color='orange',label='profit') # set the x-axis label
plt.xlabel('Month') # set the x-axis label
plt.ylabel('Amount') # set the y-axis label
plt.title('Monthly Sales and profit') # set the title of the chart
plt.xticks(rotation=45) # Rotate the x-axis labels for better readability
plt.grid(axis='y', linestyle='--',alpha=0.7) # Add horizontal gridlines with a dashed line style
plt.legend() # Add a legend

# Annotate points
for i in range(len(df)):
    plt.text(df['Month'][i],df['Sales'][i]+20,df['Sales'][i], ha='center', va='bottom') # Annotate each sales point
    plt.text(df['Month'][i],df['profit'][i]+5,df['profit'][i], ha='center', va='bottom') # Annotate each profit point

plt.show() # Display the line chart
```
